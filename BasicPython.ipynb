#### Classes can be useful for more complex tasks

```python
# A class can contain data and methods (functions).
# Again indentation is needed.

class Circle:
    def __init__(self, r, center): # Define how to initialize a Circle object
        self.r = r # Radius
        self.center = center # 2D coordinates of the center
    
    def translate(self,v_xy):
        self.center += v_xy
        
    def get_radius(self):
        return self.r
    
    def get_center(self):
        return self.center
    
    def set_radius(self, r):
        self.r = r
    
    def set_center(self, center):
        self.center = center
```

```python
mycircle = Circle(1.0, np.array([1.2,4.0]))
print(mycircle.get_center())
```

```python
mycircle.translate(np.array([-1.0,3.0]))
print(mycircle.get_center())
```

How to fit functions easily https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.curve_fit.html

```python
from scipy.optimize import curve_fit

# --- data ---
rng = np.random.default_rng(12)
x = rng.normal(loc=0, scale=1, size=1000)

meanvalue = np.mean(x)
stdvalue = np.std(x)
print(f"Mean={meanvalue:.4f}, Stand. dev={stdvalue:.4f}")

# --- histogram as fit target ---
counts, bin_edges = np.histogram(x, bins=50, density=True)
bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])

#print(np.shape(counts), np.shape(bin_edges))
#plt.hist(x, bins=20, density=True)
#plt.scatter(bin_edges, counts)
plt.scatter(bin_centers, counts)
plt.xlabel('x')
plt.ylabel('Prob density')
plt.show()
```

```python
# --- Gaussian model ---
def gauss(x, A, mu, sigma):
    return A * np.exp(-(x - mu)**2 / (2 * sigma**2))

# fit
popt, pcov = curve_fit(gauss, bin_centers, counts)
A_fit, mu_fit, sigma_fit = popt
err_A, err_mu, err_sigma = np.sqrt(np.diag(pcov))

print(f"Fit parameters:")
print(f"A     = {A_fit:.4f} ± {err_A:.4f}")
print(f"mu    = {mu_fit:.4f} ± {err_mu:.4f}")
print(f"sigma = {sigma_fit:.4f} ± {err_sigma:.4f}")

# --- plot ---
plt.figure(figsize=(7,5))
plt.hist(x, bins=50, density=True, alpha=0.6, color='skyblue', edgecolor='black', label='Samples')

xx = np.linspace(bin_edges[0], bin_edges[-1], 500)
plt.plot(xx, gauss(xx, *popt), 'r-', lw=2, label=fr'curve_fit Gaussian ($\mu={mu_fit:.3f}, \sigma={sigma_fit:.3f}$)')

plt.xlabel('x')
plt.ylabel('Probability density')
plt.legend()
plt.tight_layout()
plt.show()
```

```python

def random_walk_1d(n_steps=1000):
    x=[0]
    for i in range(n_steps):
        cointoss=np.random.uniform(0, 1)
        if cointoss>=0.5:
            xnew=x[-1]+1
        else:
            xnew=x[-1]-1
        x.append(xnew)
    timeaxis=np.arange(len(x))
    return timeaxis, x
```

```python
time, positions=random_walk_1d(1000)

plt.plot(time, positions)
plt.xlabel('Steps')
plt.ylabel('Position')
plt.show()
```

```python
# Writing
def write_to_file(particles, file):
    N=len(particles)
    with open(file, 'w') as f:
        f.write('Header \n')
        f.write('timestep position \n')
        for i in range(N):
            line=f'{i} {particles[i]} \n'
            f.write(line)
        f.close()
        
```

```python
write_to_file(positions, 'randomwalk.txt')
```

```python
# Opening
def read_from_file(file):
    particles = []
    timesteps = []

    with open(file, 'r') as f:
        lines = f.readlines()

    # skip first two header lines
    for line in lines[2:]:
        parts = line.strip().split()
        if len(parts) < 2:
            continue
        t = int(parts[0])
        x = float(parts[1])
        timesteps.append(t)
        particles.append(x)

    return timesteps, particles
    
```

```python
timesteps, particles=read_from_file('randomwalk.txt')
plt.plot(timesteps, particles)
plt.xlabel('Steps')
plt.ylabel('Positions')
plt.show()
```
